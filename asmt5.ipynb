```python
import matplotlib.pyplot as plt
import pandas as pd

# import output.csv with labeled columns
columns_names = ["title", "price", "currency", "total sold", "condition", "color", "theme", "features", "url"]
df = pd.read_csv('output.csv', names = columns_names)

# discretize price as integer instead of float (no fractions of a penny)
df.price = df.price.str.replace(',|\.', '', regex=True).astype(int)

# sold listings => at least one shirt sold, quantity not present on listings for single shirts
df['total sold'] = df['total sold'].str.replace(',', '', regex=True).astype(int)
df['total sold'].replace(0, 1, inplace=True)

# scraper issues to fix on subsequent passes
# remove listings with zero price, associated listing URL had a nonzero closing price
df = df[df.price > 0]
# remove listings with null condition, scraper only looks at 'Item Specifics', not in summary
df.dropna(subset=['condition'], inplace=True)

# listing URLs with NaN color had multiple colors available
df.color.fillna('Multiple', inplace=True)

df = df.reset_index(drop=True)

print(df)
```
